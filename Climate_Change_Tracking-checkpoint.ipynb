```python
import pandas as pd
df_global_temp = pd.read_csv("annual_csv.csv")
print(df_global_temp.head())
```

```python
df_global_temp = df_global_temp[df_global_temp['Source'] == 'GISTEMP'].reset_index()[["Source", "Year", "Mean"]]
print(df_global_temp.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
sns.lineplot(x = df_global_temp['Year'], y = df_global_temp['Mean'])
plt.ylabel("Mean")
plt.title("Average Global Mean Temperature")
plt.savefig('Average Global Mean Temperature.png', dpi=1200)
```

```python
df_sea_level = pd.read_csv("epa-sea-level_csv.csv")
print(df_sea_level.head())
```

```python
sns.set()
sns.lineplot(x = df_sea_level['Year'], y = df_sea_level['CSIRO Adjusted Sea Level'])
sns.lineplot(x= df_sea_level['Year'], y = df_sea_level['NOAA Adjusted Sea Level'])
plt.ylabel("Adjusted Sea Level")
plt.legend(['CSIRO Adjusted Sea Level', 'NOAA Adjusted Sea Level'])
plt.title("Global Average Absolute Sea Level Change")
plt.savefig('Global Average Absolute Sea Level Change.png', dpi=1200)
```
